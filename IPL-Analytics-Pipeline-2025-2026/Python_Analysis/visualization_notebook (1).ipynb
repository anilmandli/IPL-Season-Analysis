```python
from google.colab import files
uploaded = files.upload()   # pick top_wickets.csv

import pandas as pd
df = pd.read_csv('heatmap_data.csv')
# df.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_data =  df[['total_runs', 'total_balls', '4s', '6s', 'strike_rate']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Between Batting Metrics')
plt.tight_layout()
plt.savefig('heatmap.png', dpi = 150)
plt.show()
```

**Insight:** Sixes and strike rate show a strong positive correlation (0.88),
while strike rate shows a negative correlation with total balls faced (-0.79).
This suggests aggressive "finisher" type batsmen who face fewer deliveries
tend to score at a much faster rate, while batsmen with high ball counts
(anchors) trade strike rate for longer innings.

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('paretochart_data.csv')
df = df.sort_values('total_runs', ascending=False).reset_index(drop=True)

df['cumulative_pct'] = df['total_runs'].cumsum() / df['total_runs'].sum() * 100

fig, ax1 = plt.subplots(figsize=(12,6))

# Bar chart - top 20 players only for readability
top20 = df.head(20)
ax1.bar(top20['player'], top20['total_runs'], color='#2196F3')
ax1.set_ylabel('Total Runs', color='#2196F3')
plt.xticks(rotation=75, ha='right')

# Line chart - cumulative %
ax2 = ax1.twinx()
ax2.plot(top20['player'], top20['cumulative_pct'].head(20), color='#FF6B6B', marker='o')
ax2.axhline(80, color='gray', linestyle='--', linewidth=1)
ax2.set_ylabel('Cumulative % of Total Runs', color='#FF6B6B')

plt.title('Pareto Analysis — Run Contribution by Top Players')
plt.tight_layout()
plt.savefig('pareto_chart.png', dpi = 150)
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()   # pick top_wickets.csv

import pandas as pd
df = pd.read_csv('bowlerscatter_data.csv')
# df.head(10)
```

```python
df = pd.read_csv('bowlerscatter_data.csv')

median_wickets = df['total_wickets'].median()
median_economy = df['economy'].median()

fig, ax = plt.subplots(figsize=(11,7))
ax.scatter(df['economy'], df['total_wickets'], alpha=0.6, s=70, color='#2196F3')

ax.axhline(median_wickets, color='gray', linestyle='--', linewidth=1)
ax.axvline(median_economy, color='gray', linestyle='--', linewidth=1)

# Label top 10 wicket takers
top20 = df.nlargest(20, 'total_wickets')
for _, row in top20.iterrows():
    ax.annotate(row['bowler'], (row['economy'], row['total_wickets']), fontsize=8)

# Quadrant labels
ax.text(df['economy'].min(), df['total_wickets'].max(), 'Elite\n(High wickets, Low economy)',
        fontsize=9, color='green', fontweight='bold', va='top')
ax.text(df['economy'].max()-3, df['total_wickets'].max(), 'Strike Bowlers\n(High wickets, Expensive)',
        fontsize=9, color='orange', fontweight='bold', va='top')
ax.text(df['economy'].min(), df['total_wickets'].min()+2, 'Containment\n(Few wickets, Economical)',
        fontsize=9, color='blue', fontweight='bold', va='bottom')
ax.text(df['economy'].max()-3, df['total_wickets'].min()+2, 'Struggling\n(Few wickets, Expensive)',
        fontsize=9, color='red', fontweight='bold', va='bottom')

ax.set_xlabel('Economy Rate')
ax.set_ylabel('Total Wickets')
ax.set_title('Bowler Effectiveness Quadrant Analysis')
plt.tight_layout()
plt.savefig('bowler_economy.png', dpi = 150)
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()   # pick top_wickets.csv

import pandas as pd
df = pd.read_csv('heatmap_data.csv')
# df.head(10)
```

```python
df = pd.read_csv('heatmap_data.csv')

median_wickets = df['total_runs'].median()
median_economy = df['strike_rate'].median()

fig, ax = plt.subplots(figsize=(11,7))
ax.scatter(df['strike_rate'], df['total_runs'], alpha=0.6, s=70, color='#2196F3')

ax.axhline(median_wickets, color='gray', linestyle='--', linewidth=1)
ax.axvline(median_economy, color='gray', linestyle='--', linewidth=1)

top10 = df.nlargest(10, 'total_runs')
for _, row in top10.iterrows():
    ax.annotate(row['player'], (row['strike_rate'], row['total_runs']), fontsize=8)

# Quadrant labels
ax.text(df['strike_rate'].max(), df['total_runs'].max(), 'Elite\n(High Runs, High Strike Rate)',
        fontsize=9, color='green', fontweight='bold', va='top')
ax.text(df['strike_rate'].min(), df['total_runs'].max(), 'Consistance Players\n(Low Strike Rate, High Runs)',
        fontsize=9, color='orange', fontweight='bold', va='top')
ax.text(df['strike_rate'].max(), df['total_runs'].min(), 'Striker Players\n(Low Runs, High Strike Rate)',
        fontsize=9, color='blue', fontweight='bold', va='top')
ax.text(df['strike_rate'].min(), df['total_runs'].min(), 'Struggling\n(Low Runs,Low strike Rate)',
        fontsize=9, color='red', fontweight='bold', va='center_baseline')

ax.set_xlabel('Strike Rate')
ax.set_ylabel('Total Runs')
ax.set_title('Batter Effectiveness Quadrant Analysis')
plt.tight_layout()
plt.savefig('batter_Strike.png', dpi = 150)
plt.show()
```
