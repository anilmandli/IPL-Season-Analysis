```python
from google.colab import files
uploaded = files.upload()   # pick top_wickets.csv

import pandas as pd
df = pd.read_csv('1_teams_win_season.csv')#1_teams_win_season.csv
# df.head(10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12,6))

b1 = plt.barh(df['team_name'],df['wins_2025'], color ='purple', label = 'Wins 2025')

b2 = plt.barh(df['team_name'],df['wins_2026'], left=df['wins_2025'], color ='brown', label = 'Wins 2026')

plt.bar_label(b1, label_type = 'center', color = 'white', fontweight = 'bold')
plt.bar_label(b2, label_type = 'center', color = 'white', fontweight = 'bold',)

plt.bar_label(b2, labels = df['total_wins'], padding = 5, fontweight = 'bold')

plt.xlabel('Number of Wins')
plt.title('Total Wins & Yearly Breakdown For Each Team — IPL 2025 & 2026')
plt.legend(loc = 'lower right')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('total_wins_stackes.png', dpi = 150)
plt.show()
```

```python
from google.colab import files
uploaded = files.upload()   # pick top_wickets.csv

import pandas as pd
df = pd.read_csv('total_per_season.csv') #file used = bat_n_ball_win2.csv
# df.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

teams = df['team']
runs_2025 = df['runs_2025']
runs_2026 = df['runs_2026']

y = np.arange(len(teams))
height = 0.35

plt.figure(figsize = (12,8))

b1 = plt.barh(y - height/2, runs_2025, height, label='2025 Runs', color='#17A2B8')

b2 = plt.barh(y + height/2, runs_2026, height, label='2026 Runs', color='#6F42C1')

plt.bar_label(b1, padding=3, fontsize=9)
plt.bar_label(b2, padding=3, fontsize=9)


plt.yticks(y, teams)
plt.xlim(0, max(max(runs_2025), max(runs_2026)) + 300)
plt.xlabel('Runs Scored')
plt.title('Season-wise Runs Comparison: IPL 2025 vs 2026')
plt.legend(loc='lower right')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
